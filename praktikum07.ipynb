1. import library

```python
import matplotlib.pyplot as plt
import numpy as np
```

2. Stacked Horizontal Bar Chart dengan Matplotlib

```python
#data penjualan
brands = ['Samsung', 'Iphone', 'Vivo', 'Oppo']
sales_2023 = [500, 700, 300, 400]
sales_2024 = [550, 750, 350, 450]
sales_2025 = [600, 800, 400, 500]

y_pos = np.arange(len(brands))

plt.barh(brands, sales_2023, color='skyblue', label='2023')
plt.barh(brands, sales_2024, left=sales_2023, color='salmon', label='2024')

plt.barh(brands, sales_2023, color='skyblue')
plt.title('Penjualan Smartphone Tahun 2023')
plt.xlabel('Jumlah Penjualan')
plt.ylabel('Merek')
plt.show()
```

3. Kustomisasi Stacked Horizontal Bar Chart

```python
# Data penjualan smartphone
brands = ['Samsung', 'Iphone', 'Vivo', 'Oppo']
sales_2023 = [500, 700, 300, 400]
sales_2024 = [550, 750, 350, 450]
sales_2025 = [600, 800, 400, 500]
colors = ['gold', 'silver', 'coral', 'lightgreen']
y_pos = np.arange(len(brands))
plt.barh(brands, sales_2023, color='lightcoral', edgecolor='black', label='2023')
plt.barh(brands, sales_2024, left=sales_2023, color='lightskyblue', edgecolor='black', label='2024')
plt.barh(brands, sales_2023, color=colors, edgecolor='black', height=0.6)
plt.title('Penjualan Smartphone Tahun 2023 (Kustom)')
plt.xlabel('Jumlah Penjualan')
plt.ylabel('Merek')
plt.grid(axis='x', linestyle='--', alpha=0.5)
plt.show()
```

4. Multiple Stacked Horizontal Bar Chart

```python
# Data penjualan smartphone
brands = ['Samsung', 'Iphone', 'Vivo', 'Oppo']
sales_2023 = [500, 700, 300, 400]
sales_2024 = [550, 750, 350, 450]
sales_2025 = [600, 800, 400, 500]

y_pos = np.arange(len(brands))

plt.barh(y_pos - 0.2, sales_2023, color='skyblue', height=0.4, label='2023')
plt.barh(y_pos + 0.2, sales_2024, color='salmon', height=0.4, label='2024')

plt.barh(y_pos, sales_2023, color='skyblue', label='2023')
plt.barh(y_pos, sales_2024, left=sales_2023, color='salmon', label='2024')
plt.barh(y_pos, sales_2025, left=np.array(sales_2023) + np.array(sales_2024), color='lime',label='2025')

plt.yticks(y_pos, brands)
plt.title('Penjualan Smartphone 2023 vs 2024')
plt.xlabel('Jumlah Penjualan')
plt.ylabel('Merek')
plt.legend()
plt.show()
```

5. Area Chart dengan Matplotlib

```python
# Data Penjualan Bulanan
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun',
          'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
shoes = [500, 600, 700, 800, 650, 700, 850, 900, 750, 800, 950, 1000]

# Membuat Area Chart Dasar
plt.figure(figsize=(10, 6))
plt.fill_between(months, shoes, color='blue', alpha=0.4, label='Shoes')
plt.title('Basic Area Chart: Monthly Sales')
plt.xlabel('Months')
plt.ylabel('Units Sold')
plt.legend()
plt.show()
```

6. Kustomisasi Area Chart

```python
# Data Penjualan Bulanan
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun',
          'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
shoes = [500, 600, 700, 800, 650, 700, 850, 900, 750, 800, 600, 1000]
sandals = [300, 350, 400, 450, 500, 300, 600, 400, 700, 750, 500, 850]

plt.figure(figsize=(10, 6))
plt.fill_between(months, shoes, color='blue', alpha=0.5, label='Shoes')
plt.fill_between(months, sandals, color='green', alpha=0.5, label='Sandals')
plt.title('Customized Area Chart: Monthly Sales')
plt.xlabel('Months')
plt.ylabel('Units Sold')
plt.xticks(rotation=45)  # Memutar label bulan
plt.grid(axis='y', linestyle='--', alpha=0.6)  # Garis grid horizontal
plt.legend()
plt.show()
```

7. Multiple Area Chart

```python
# Data Penjualan Bulanan
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun',
          'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
shoes = [500, 600, 700, 800, 650, 700, 850, 900, 750, 800, 950, 1000]
sandals = [300, 350, 400, 450, 500, 550, 600, 650, 700, 750, 800, 850]
socks = [200, 250, 300, 350, 300, 400, 450, 500, 600, 700, 750, 800]

# Multiple Area Chart
plt.figure(figsize=(10, 6))
plt.fill_between(months, shoes, color='blue', alpha=0.4, label='Shoes')
plt.fill_between(
    months,
    np.add(shoes, sandals),
    shoes,
    color='green',
    alpha=0.4,
    label='Sandals'
)
plt.fill_between(
    months,
    np.add(shoes, np.add(sandals, socks)),
    np.add(shoes, sandals),
    color='orange',
    alpha=0.4,
    label='Socks'
)

plt.title('Multiple Area Chart: Monthly Sales')
plt.xlabel('Months')
plt.ylabel('Units Sold')
plt.legend()
plt.show()
```
